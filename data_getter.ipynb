### Broad Index Data

Let’s look at the live index data first:

```python
from jugaad_data.nse import stock_df
from datetime import date


sbi_stock_df = stock_df(symbol='SBIN', 
                                from_date=date(2000,1,1), 
                                to_date=date(2024,1,1), 
                                series="EQ")
sbi_stock_df
```

```python
list_of_banks = ['SBIN', 'HDFC', 'AXISBANK', 'PNB', 'PSB', 'IDFCFIRSTB', 'BANKBARODA', 'KOTAKBANK', 'ICICIBANK']
```

```python
for bank_symbol in list_of_banks:
    print(f'{bank_symbol}'+'_'+'df.csv')
```

```python

def get_bank_data(bank_symbol):
    '''
    pass both fromdate & to_date in yyyy,mm,dd format only
    '''
    df = stock_df(symbol=str(bank_symbol),
                  from_date=date(2000,1,1),
                  to_date=date(2024,7,30),
                  series="EQ")
    df.to_csv(f'{bank_symbol}'+'_'+'df.csv')
```

```python
for bank in list_of_banks:
    get_bank_data(bank)
```
